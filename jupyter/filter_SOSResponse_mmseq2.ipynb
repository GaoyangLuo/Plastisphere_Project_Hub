```python
import os
import pandas as pd

# Step 1: 读取 RPKG 文件并构建字典（仅包含 'Name' 和 'RPKG' 列）
def build_rpkg_dict_for_file(rpkg_file):
    try:
        # 只读取 'Name' 和 'RPKG' 列，并将其转换为字典
        rpkg_df = pd.read_csv(rpkg_file, usecols=['#Name', 'RPKG'])
        rpkg_dict = dict(zip(rpkg_df['#Name'], rpkg_df['RPKG']))
        return rpkg_dict
    except Exception as e:
        print(f"Error processing RPKG file {rpkg_file}: {e}")
        return {}

# Step 2: 读取 m8 文件并过滤
def filter_m8_file(m8_file):
    # 检查文件是否为空
    if os.stat(m8_file).st_size == 0:
        print(f"File {m8_file} is empty. Returning sum as 0.")
        return None

    try:
        df = pd.read_csv(m8_file, sep="\t", header=None)
        df.columns = ['id', 'sub_id', 'identity', 'alignLen', 'mismat', 'gapOpens', 'qStart', 'qEnd', 'sStart', 'sEnd', 'eval', 'bit']

        # 过滤条件：identity > 0.7 且 alignLen > 40
        filter_df = df[(df['identity'] > 0.7) & (df['alignLen'] > 40)]

        # 获取过滤后的 contig id 列表
        contig_list = list(set(filter_df['id']))
        return contig_list
    except pd.errors.EmptyDataError:
        print(f"File {m8_file} has no data to parse. Returning sum as 0.")
        return None

# Step 3: 根据 contig_list 查询 RPKG 字典
def get_rpkg_data_from_dict(contig_list, rpkg_dict, file_name):
    rpkg_data = []
    
    # 遍历 contig_list，使用字典进行查询
    for contig_id in contig_list:
        # 去掉最后一个下划线及后面的部分，得到形如 k141_1350569_1 的格式
        truncated_id = "_".join(contig_id.split("_")[:-1])
        
        # 在字典中查找处理过的 id
        if truncated_id in rpkg_dict:
            rpkg_data.append({
                'id': contig_id,  # 保留原始的 contig_id
                'RPKG': rpkg_dict[truncated_id],
                'file': file_name,  # 添加文件名
                'sample_type': extract_sample_type(file_name)  # 添加样本种类
            })
    
    return rpkg_data

def extract_sample_type(file_name):
    # 假设文件名格式为 "S0PCL.contigs_gpx.m8"
    # 提取文件名中第一个 '.' 前的部分，作为 sample_type
    return file_name.split('.')[0]

# Step 5: 计算每个样本的 RPKG 总和和平均值
def calculate_rpkg_sum(rpkg_data):
    # 转换为 DataFrame 以便进行计算
    df = pd.DataFrame(rpkg_data)
    
    # 按照 file 计算 RPKG 的总和和平均值
    rpkg_sum_mean = df.groupby('file').agg(
        RPKG_sum=('RPKG', 'sum'),     # 计算 RPKG 总和
        RPKG_mean=('RPKG', 'mean')    # 计算 RPKG 平均值
    ).reset_index()
    
    return rpkg_sum_mean

# Step 6: 保存两个文件（更新后的版本）
def save_files(output_dir, base_name, identifier, rpkg_data_list, rpkg_sum_mean_df):
    # 创建输出目录（如果不存在）
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # 保存匹配的 RPKG 数据到 CSV 文件
    if rpkg_data_list:
        rpkg_data_df = pd.DataFrame(rpkg_data_list)
        data_output_path = os.path.join(output_dir, f"{base_name}_{identifier}_matched_rpkg_data.csv")
        rpkg_data_df.to_csv(data_output_path, index=False)
        print(f"Saved matched RPKG data to {data_output_path}")
    
    # 保存 RPKG 总和和平均值到 CSV 文件
    sum_output_path = os.path.join(output_dir, f"{base_name}_{identifier}_rpkg_sum.csv")
    rpkg_sum_mean_df.to_csv(sum_output_path, index=False)
    print(f"Saved RPKG sum and mean to {sum_output_path}")

# Step 8: 主函数，执行整个流程（保持不变）
def main():
    # m8 文件路径
    m8_dir = "/storage/jufengLab/luogaoyang/metagenome_project/DSR/4_annotation/ROS_gene_mmseq_out"
    
    # RPKG 文件夹路径
    rpkg_dir = "/storage/jufengLab/luogaoyang/metagenome_project/DSR/PRKG"
    
    # 输出文件夹路径
    output_dir = "/storage/jufengLab/luogaoyang/metagenome_project/DSR/4_annotation/ROS_gene_mmseq_out/rpkg_cal"

    # Step 1: 获取所有 m8 文件
    for m8_file in os.listdir(m8_dir):
        if m8_file.endswith(".m8"):
            m8_path = os.path.join(m8_dir, m8_file)
            print(f"Processing {m8_file}...")
            
            # Step 2: 过滤 m8 文件，获取 contig 列表
            contig_list = filter_m8_file(m8_path)
            
            # Step 3: 解析文件名，提取样本基本名和标识符
            base_name, identifier = parse_file_name(m8_file)
            
            # 如果文件为空或没有匹配的 contig，保存 RPKG_sum = 0 的文件
            if contig_list is None or not contig_list:
                rpkg_sum_df = pd.DataFrame({'file': [f"{base_name}_{identifier}"], 'RPKG_sum': [0], 'RPKG_mean': [0]})
                save_files(output_dir, base_name, identifier, [], rpkg_sum_df)
                continue

            # Step 4: 查找对应的 RPKG 文件
            rpkg_file = os.path.join(rpkg_dir, f"{base_name}_RPKG.csv")
            if not os.path.exists(rpkg_file):
                print(f"RPKG file {rpkg_file} not found. Skipping.")
                continue

            # Step 5: 构建 RPKG 字典，使用 pandas 读取 'Name' 和 'RPKG' 列
            rpkg_dict = build_rpkg_dict_for_file(rpkg_file)

            # Step 6: 根据 contig_list 从字典中查找 RPKG 数据
            rpkg_data = get_rpkg_data_from_dict(contig_list, rpkg_dict, f"{base_name}_{identifier}")
            
            # 如果有匹配数据，处理并保存文件
            if rpkg_data:
                # Step 7: 计算 RPKG 总和和平均值
                rpkg_sum_df = calculate_rpkg_sum(rpkg_data)
                
                # Step 8: 保存匹配的 RPKG 数据和 RPKG 总和文件
                save_files(output_dir, base_name, identifier, rpkg_data, rpkg_sum_df)
            else:
                print(f"No matching RPKG data for {m8_file}")

if __name__ == "__main__":
    main()
```

```python
import os
import pandas as pd

# Step 1: 定义处理函数
def merge_rpkg_sum_files_by_gene(input_dir, output_dir):
    # 创建输出目录（如果不存在）
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # 获取所有的文件列表
    files = os.listdir(input_dir)
    
    # 用字典存储每个基因对应的文件数据
    gene_data = {}

    # Step 2: 遍历目录中的文件
    for file in files:
        if file.endswith('_rpkg_sum.csv'):
            file_path = os.path.join(input_dir, file)
            df = pd.read_csv(file_path)
            
            # 提取基因名称，比如 '_lex_', '_gpx_'，通过文件名识别基因类型
            gene = file.split('_')[1]  # 假设基因名是文件名的第二部分
            
            # 如果字典中该基因还没有数据，初始化一个空的 DataFrame 列表
            if gene not in gene_data:
                gene_data[gene] = []
                
            # 将当前文件的数据添加到对应基因的列表中
            gene_data[gene].append(df)

    # Step 3: 对每个基因的数据进行汇总
    for gene, dataframes in gene_data.items():
        # 将所有 DataFrame 合并为一个
        merged_df = pd.concat(dataframes, ignore_index=True)
        
        # 保存为新的文件，文件名格式为 "{gene}_rpkg_sum_merged.csv"
        output_file = os.path.join(output_dir, f"{gene}_rpkg_sum_merged.csv")
        merged_df.to_csv(output_file, index=False)
        print(f"Saved merged {gene} RPKG sum file to {output_file}")

# Step 4: 调用函数
if __name__ == "__main__":
    # 输入目录包含所有 RPKG_sum 文件
    input_dir = "/storage/jufengLab/luogaoyang/metagenome_project/DSR/4_annotation/ROS_gene_mmseq_out/rpkg_cal"  # 你需要修改为实际的路径
    
    # 输出目录保存汇总后的文件
    output_dir = "/storage/jufengLab/luogaoyang/metagenome_project/DSR/4_annotation/ROS_gene_mmseq_out/rpkg_cal/merged_output"  # 你需要修改为实际的路径

    # 调用函数，生成基因汇总文件
    merge_rpkg_sum_files_by_gene(input_dir, output_dir)
```
